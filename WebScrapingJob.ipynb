```python
# importing libraries and fetching or scraping the data
import requests
from bs4 import BeautifulSoup

#url = input('Enter the link:')
url = "https://realpython.github.io/fake-jobs/"

page = requests.get(url)

if page.status_code == 200:
  soup = BeautifulSoup(page.content, "html.parser") # .content instead of .text to avoid problems with character encoding.
else:
  print("Failed to retrieve the webpage. Status code:", page.status_code)
```

```python
results = soup.find(id="ResultsContainer")

#print(results.text) #print the raw results
#print(results.prettify()) #print structured results, removing extra spaces, identations, etc

# After inspecting the website we found our target, h2 (title), h3 (company name) and p (location):
# Specifically, the job Python Developer
#<h2 class="title is-5">Senior Python Developer</h2>
```

```python
python_jobs = results.find_all("h2", string=lambda text: "python" in text.lower())

print("Found", len(python_jobs), "Python-related jobs...\n")

python_job_cards = [
    h2_element.parent.parent.parent for h2_element in python_jobs
]

for job_card in python_job_cards:
    title_element = job_card.find("h2", class_="title")
    company_element = job_card.find("h3", class_="company")
    location_element = job_card.find("p", class_="location")
    print(title_element.text.strip())
    print(company_element.text.strip())
    print(location_element.text.strip())
    link_url = job_card.find_all("a")[1]["href"]
    print(f"Apply here: {link_url}\n")
```
